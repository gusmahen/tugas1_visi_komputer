# Intensitas tiap titik

```python
import cv2
from google.colab.patches import cv2_imshow # import imshow khusus untuk Colab

# Mengambil gambar grayscale
image_path = "/content/angsa.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Membaca gambar dalam mode grayscale

if img is not None:
    # Resize gambar menjadi 100x100 piksel
    img = cv2.resize(img, (100, 100))  # Mengubah ukuran gambar menjadi 100x100

    height, width = img.shape
    for x in range(height):
        for y in range(width):
            # Nilai intensitas langsung dari citra grayscale
            intensity = img[x, y]
            print(f"Nilai intensitas pada {x},{y}: {intensity}")

    # Menampilkan gambar
    cv2_imshow(img)
```

# Rotasi 90 drajat searah jarum jam

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # import imshow khusus untuk Colab

# Mengambil gambar grayscale
image_path = "/content/angsa.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Membaca gambar dalam mode grayscale
print("Gambar Sebelum Rotasi")
cv2_imshow(img)

if img is not None:
    height, width = img.shape

    # Rotasi 90 Derajat Searah Jarum Jam
    rotated_img = np.zeros((width, height), dtype=np.uint8)  # Membuat array kosong untuk gambar yang dirotasi

    for x in range(height): # baris pertama dari gambar asli menjadi kolom terakhir di gambar yang dirotasi.
     for y in range(width):
            rotated_img[y, height - 1 - x] = img[x, y] # y pada gambar asli menjadi y pada gambar yang dirotasi, nilai x bertransformasi menjadi height - 1 - x, yang memindahkan baris ke kolom yang sesuai setelah rotasi.

    print("Gambar Setelah Rotasi 90 Derajat Searah Jarum Jam:")
    cv2_imshow(rotated_img)
else:
    print("Error: Gambar tidak ditemukan atau format salah")
```

# Flip Horizontal

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # import imshow khusus untuk Colab

# Mengambil gambar grayscale
image_path = "/content/angsa.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Membaca gambar dalam mode grayscale
print("Gambar Sebelum flip horizontal")
cv2_imshow(img)

if img is not None:
    height, width = img.shape

    # Flip Horizontal
    flipped_img_horizontal = np.zeros_like(img)  # Membuat array kosong untuk gambar yang dibalik horizontal
    for x in range(height):
        for y in range(width):
            flipped_img_horizontal[x, width - 1 - y] = img[x, y]

    print("Gambar Setelah Flip Horizontal:")
    cv2_imshow(flipped_img_horizontal)
else:
    print("Error: Gambar tidak ditemukan atau format salah")
```

# Flip Vertikal

# Flip Horizontal dan Vertikal

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # import imshow khusus untuk Colab

# Mengambil gambar grayscale
image_path = "/content/angsa.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Membaca gambar dalam mode grayscale
print("Gambar Sebelum flip horizontal & vertikal")
cv2_imshow(img)

if img is not None:
    height, width = img.shape

    # Flip Horizontal dan Vertikal
    flipped_img_both = np.zeros_like(img)  # Membuat array kosong untuk gambar yang dibalik horizontal dan vertikal
    for x in range(height):
        for y in range(width):
            flipped_img_both[height - 1 - x, width - 1 - y] = img[x, y]

    print("Gambar Setelah Flip Horizontal dan Vertikal:")
    cv2_imshow(flipped_img_both)
else:
    print("Error: Gambar tidak ditemukan atau format salah")
```
